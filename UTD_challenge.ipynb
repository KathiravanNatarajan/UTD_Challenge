```python
import numpy as np 
import pandas as pd 
train_Set = pd.read_excel('Training Dataset.xlsx')
train_Set.head()
```

```python
list(train_Set)
```

```python
# Feature Engineering 
new_df = pd.get_dummies(train_Set, columns=['Education Level','Age Range','Gender'])
new_df = new_df.fillna(0)
new_df.head()
```

```python
train_X = new_df.drop(['Total','Employment Status'],axis=1)
train_y = new_df['Total']
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit(train_X, train_y)
```

```python
# Test Set 
Test_Set = pd.read_excel('Test Dataset.xlsx')
test_df = pd.get_dummies(Test_Set, columns=['Education Level','Age Range','Gender'])
test_df = test_df.fillna(0)
test_X = test_df.drop(['Total', 'Employment Status'],axis=1)
test_y = test_df['Total']
```

```python
# Make predictions using the testing set
from sklearn.metrics import precision_score
test_pred = regr.predict(test_X)
print(precision_score(test_y, test_pred, average='weighted'))
```
